```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

```python
df1 = pd.read_csv('olist_data/olist_sellers_dataset.csv')
df1.sample(10)
```

<hr />

```python
df2 = pd.read_csv('olist_data/olist_products_dataset.csv')
df2.sample(10)
```

```python
df2.info()
```

<hr />

```python
vendas_qtd = df2.product_category_name.value_counts()
vendas_qtd.head(15)
```

```python
a = df2.groupby('product_category_name').product_photos_qty.agg([len, min, max, np.mean, np.median, np.std])
a
```

```python
fotos_qtd = df2.groupby('product_category_name').product_photos_qty.mean()
fotos_qtd
```

```python
sns.scatterplot(x=vendas_qtd, y=fotos_qtd)
sns.regplot(x=vendas_qtd, y=fotos_qtd)
```

## A tendência de produtos com mais vendas é ter menos fotos

```python
b = df2.groupby('product_category_name').product_description_lenght.agg([len, min, max, np.mean, np.median, np.std])
b
```

```python
descricao_qtd = df2.groupby('product_category_name').product_description_lenght.median()
descricao_qtd
```

```python
sns.scatterplot(x=vendas_qtd, y=descricao_qtd)
sns.regplot(x=vendas_qtd, y=descricao_qtd)
```

## A tendência de produtos com mais vendas é ter uma descrição menor
